```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the cleaned data
data = pd.read_csv("/content/cleaned_data.csv")
```

```python
#Pie Chart of Lunch Type
plt.figure(figsize=(8, 6))
data['lunch'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Lunch Type')
plt.ylabel('')
plt.savefig('Distribution_of_Lunch_Type.png')
plt.show()
```

```python
#Violin Plot of Writing Scores by Lunch Type
plt.figure(figsize=(10, 6))
sns.violinplot(x='lunch', y='writing score', data=data)
plt.title('Writing Scores by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Writing Score')
plt.savefig('Writing_Scores_by_Lunch_Type.png')
plt.show()
```

```python
#Heatmap of Correlations
correlation_matrix = data[['math score', 'reading score', 'writing score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Scores')
plt.savefig('Correlation_Heatmap_of_Scores.png')
plt.show()
```

```python
#Stacked Bar Plot of Gender and Lunch Type
gender_lunch_counts = data.groupby(['gender', 'lunch']).size().unstack(fill_value=0)
gender_lunch_counts.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Stacked Bar Plot of Gender and Lunch Type')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Lunch Type')
plt.savefig('Stacked_Bar_Plot_of_Gender_and_Lunch_Type.png')
plt.show()
```

```python
# 11. Box Plot of Reading Scores by Education Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='reading score', data=data)
plt.title('Reading Scores by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Reading Score')
plt.xticks(rotation=45)
plt.savefig('Reading_Scores_by_Education_Level.png')
plt.show()
```
