```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the cleaned data
cleaned_data = pd.read_csv("/content/cleaned_data.csv")
```

```python
#Histogram of Age
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['Age'], bins=10, kde=True)
plt.title("Histogram of Age")
plt.savefig("histogram_age.png")
plt.show()
```

```python
#Scatter plot of Grip Strength vs Age
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='grip_strength_kilograms', data=cleaned_data)
plt.title("Scatter plot of Grip strength vs. Age")
plt.savefig("scatter_age_grip.png")
plt.show()
```
